### 1.3 reaname dataframe cols

```python
data_ratings = data_ratings.rename(columns={'User-ID': 'UserID', 'ISBN':'BookID', 'Book-Rating':'Rating'})
```

```python
data_ratings.head()
```

### 1.4 Data Selection - Define hyperparameters
such as min_book_rating, no_of_data_row

```python

## Min Movie Rating range [0, 5]
min_book_rating = 3

## No_of_data_row == userID
no_of_data_row = 2000
```

```python
all_ratings = data_ratings.copy(deep=True)
```

```python
all_ratings.head()
```

```python
## Choosing >3 ratings only, as 'Favorable Ratings'
all_ratings["Favorable"] = all_ratings["Rating"] > min_book_rating
```

```python
## for how many users? This is a point to modify dataset size

ratings = all_ratings[all_ratings['UserID'].isin(range(no_of_data_row))]
```

```python
favorable_ratings = ratings[ratings["Favorable"]]
```

```python
print("all_ratings len: ", len(all_ratings))
print("ratings len: ", len(ratings))
print("favorable_ratings len: ", len(favorable_ratings))
```

### 1.5 All Books Group by User -- Itemset
movie ids grouped by userid. this is the itemset

```python
favorable_reviews_by_users = dict((k, frozenset(v.values)) for k, v in favorable_ratings.groupby("UserID")["BookID"])
```

```python
data_list = [list(v.values) for k, v in favorable_ratings.groupby("UserID")["BookID"] ]
```

### 1.6 Getting Book Names from BookID

```python
# movie_name_filename = 'Books.csv'
# movie_name_data = pd.read_csv(movie_name_filename, encoding = "mac-roman")
book_name_data = data_books.copy()
book_name_data.drop(columns=['Book-Author','Year-Of-Publication','Publisher', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L'], inplace=True)
book_name_data.rename(columns={'ISBN':'BookID', 'Book-Title': 'Title'}, inplace=True)
```

```python
book_name_data.head()
```

```python
def get_book_name(book_id):
    title_object = book_name_data[book_name_data["BookID"] == book_id]["Title"]
    title = title_object.values[0]
    return title
```
